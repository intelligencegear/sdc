```python
from keras.models import load_model
from IPython.display import Image, display, SVG
from keras.utils.visualize_util import model_to_dot
from keras.utils.visualize_util import plot

best_model_filepath = './nvidia_model.h5'
model = load_model(best_model_filepath)

# Show the model in ipython notebook
figure = SVG(model_to_dot(model, show_shapes=True).create(prog='dot', format='svg'))
display(figure)
# Save the model as png file
plot(model, to_file='nvidia_model.png', show_shapes=True)
```
